```python
#chrome창(웹드라이버) 열기 
browser = webdriver.Chrome("C:/Users/debbie/Downloads/chromedriver.exe") 

#사진 저장할 폴더 경로 

# 홈페이지 접속하기 
name = '아이유'
url = 'https://www.google.co.in/search?q=' + name + '&tbm=isch'
browser.get(url) #접속
time.sleep(0.5)

# 도구 버튼 클릭 
dogoo_button = '//*[@id="yDmH0d"]/div[2]/c-wiz/div[1]/div/div[1]/div[2]/div[2]/div'
browser.find_element_by_xpath(dogoo_button).click()

# 크기 버튼 클릭
size_button = '//*[@id="yDmH0d"]/div[2]/c-wiz/div[2]/div[2]/c-wiz[1]/div/div/div[1]/div/div[1]/div/div[1]'
browser.find_element_by_xpath(size_button).click()

# 큰 사진만 클릭
large_xpath = '//a[@aria-label="큼"]'
browser.find_element_by_xpath(large_xpath).click()
```

```python
imgs = browser.find_elements_by_css_selector(".rg_i.Q4LuWd") # 작은 이미지들
count = 1

for img in imgs:
    try:
        img.click() # 개별 이미지 클릭
        time.sleep(1)
        photo_elements = browser.find_elements_by_tag_name('img')

        for idx, ph in enumerate(photo_elements):
            photo_url = ph.get_attribute('src')
            photo_alt = ph.get_attribute('alt')
            photo_class = ph.get_attribute('class')
            
            if photo_url and photo_alt and photo_class != 'dwrY3b':
                photo_name = f'{name}_{idx}.jpg'
                path = 'C:\\Users\\debbie\\PycharmProjects\\imgs\\' + name + '\\'
                os.makedirs(path, exist_ok=True)
                urllib.request.urlretrieve(photo_url, path + photo_name)
                print(f'{photo_name} 다운로드가 완료되었습니다.')
            
            count += 1
            if count == 3:
                break

    except:
        pass

browser.close()
```
